```python
# Filter dataframe utama untuk mendapatkan hanya konten yang berhubungan dengan Indonesia
df_indo = df[df['country'].str.contains('Indonesia', na=False)].copy()

print(f"Analisis akan difokuskan pada {len(df_indo)} judul konten Indonesia.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur style visualisasi
sns.set_style('darkgrid')

# Analisis Genre
genre_indo = df_indo['listed_in'].str.split(', ').explode().value_counts()

plt.figure(figsize=(10, 8))
sns.barplot(y=genre_indo.head(10).index, x=genre_indo.head(10).values, palette='viridis')
plt.title('Top 10 Genre Konten Indonesia di Netflix (Hingga 2021)')
plt.xlabel('Jumlah Judul')
plt.show()
```

# Analisis Genre

Melihat dari top Genre terlihat bahawa genre drama dan romantic di tahun 2022 maka peluang untuk genre sangatlah bagus, makanya sekarang di tahun 2025 sangat banyak production house yang membuat genre Horor.

```python
# Analisis Aktor
# Kita perlu drop baris yang data 'cast'-nya kosong dulu
df_indo.dropna(subset=['cast'], inplace=True)
aktor_indo = df_indo['cast'].str.split(', ').explode().value_counts()

plt.figure(figsize=(10, 8))
sns.barplot(y=aktor_indo.head(10).index, x=aktor_indo.head(10).values, palette='plasma')
plt.title('Top 10 Aktor/Aktris di Konten Indonesia Netflix')
plt.xlabel('Jumlah Penampilan')
plt.show()
```

# Analisis Aktor

Aktor seperti Reza Rahadian Mudy Koesnaedi memiliki rekam jejak yang kuat di konten yang diakuisisi Netflix, menjadikan mereka pilihan 'aman' untuk meningkatkan nilai komersial sebuah proyek di mata platform streaming.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

# Analisis Tren Rilis Tahunan
plt.figure(figsize=(12, 6))

# Kita hitung jumlah konten indonesia per tahun rilis
rilis_per_tahun = df_indo['release_year'].value_counts().sort_index()

# Kita fokus pada tahun 2000 ke atas agar grafiknya lebih jelas
rilis_per_tahun[rilis_per_tahun.index >= 2000].plot(kind='line', marker='o')

# Set x-axis ticks to display integers
plt.gca().xaxis.set_major_locator(mticker.MaxNLocator(integer=True))

plt.title('Pertumbuhan Jumlah Konten Indonesia di Netflix Berdasarkan Tahun Rilis')
plt.xlabel('Tahun Rilis')
plt.ylabel('Jumlah Judul')
plt.grid(True)
plt.show()
```

```python
# Install library AI
!pip install transformers -q

from transformers import pipeline

# Siapkan model AI
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Siapkan data sample: 5 deskripsi film horor Indonesia, excluding 'Trese'
df_horor_indo = df_indo[
    (df_indo['listed_in'].str.contains('Horror')) &
    (df_indo['title'] != 'Trese')
].head(5)
descriptions = df_horor_indo['description'].tolist()

# Siapkan label/kategori yang kita inginkan
candidate_labels = ['horor psikologis', 'horor folklor/mitos', 'horor slasher', 'komedi horor']

# Lakukan klasifikasi
results = classifier(descriptions, candidate_labels)

# Tampilkan hasilnya dengan rapi
for i, result in enumerate(results):
    title = df_horor_indo.iloc[i]['title']
    top_genre = result['labels'][0]
    score = result['scores'][0]
    print(f"Film: {title}\nDeskripsi: {descriptions[i]}\nAI mendeteksi sebagai: {top_genre} (Skor: {score:.2f})\n---")
```
